Zunächst führen wir eine logistiche Regression in R durch. Dafür verwenden wir den "iris" Datensatz, in dem es um Blüten geht.
Wir laden den Datensatz, verschaffen uns einen Überblick und schauen uns die Response-Variable "Species" (welche Pflanzenart) an.

```r
data(iris)
head(iris)
plot(iris$Species)
```

Unsere Response hat drei mögliche Werte. Die logistische Regression ist aber nur für binäre Variablen geeignet. Wir verwenden daher nur die ersten 100 Werte.

```r
ir_data <- iris[1:100,]
```

Als nächstes wollen wir eine logistische Regression fitten, die "Species" vorhersagt.

```r
fit <- glm(ir_data$Species ~ ir_data$Sepal.Length, family = "binomial")
summary(fit)
```

Der Datensatz ist relativ klein und durch die Auswahl mehrerer unabhängiger Variablen kommen wir schon in der Situation, dass die Daten zum Teil eindeutig vorhergesagt werden können. Daher die Warnung.

Nun wollen wir uns anschauen, welche Klassen unser Model "fit" vorhersagt. Dafür benutzen wir "predict" und geben einen neuen Wert für Sepal.Length ein.

```r
newdata <- data.frame(x=ir_data$Sepal.Length)
predicted_val <- predict(fit, newdata, type="response")
prediction <- data.frame(ir_data$Sepal.Length, ir_data$Species,predicted_val)
prediction
```

```r
library(ggplot2)
qplot(prediction[,1], round(prediction[,3]), col=prediction[,2], xlab = 'Sepal Length', ylab = 'Prediction using Logistic Reg.')
```

Als nächstes wollen wir eine Kreuztabelle für die Vorhersagen erstellen. Dafür runden wir einfach die ermittelten Wahrscheinlichkeiten. (>=.05 = 1)

```r
classDF <- data.frame(response = ir_data$Species, predicted = round(fitted(fit),0))

t <-xtabs(~ predicted + response, data = classDF)[,-3]
t
# Die Variable enthält noch die dritte Klasse, auch wenn es dafür keine Beobachtung gibt. Das [,-3] schmeißt einfach die dritte Spalte, die wir nicht brauchen, raus.
```

Mit unserer Tabelle t können wir jetzt die Accuaracy berechnen. Also: Wie viel Prozent der Beobachtungen wurden richtig vorhergesagt.

```r
(t[1,1] + t[2,2])/100 # 100 ist die Anzahl der Beobachtungen.
```
